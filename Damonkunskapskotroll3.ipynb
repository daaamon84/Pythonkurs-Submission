```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Ls in data
housing = pd.read_csv("housing.csv")

# Skapa en scatter-plot baserad p longitude och latitude
plt.figure(figsize=(10, 8))
plt.scatter(housing['longitude'], housing['latitude'], alpha=0.4, 
            c=housing['median_house_value'], cmap='viridis', s=10)
plt.colorbar(label='Median House Value')
plt.title('Geographical Distribution of Houses')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
missing_values = housing.isnull().sum()

# Skriv ut resultatet
print("Saknade vrden i varje kolumn:")
print(missing_values)
```

```python
import pandas as pd

housing = pd.read_csv("housing.csv")

# Berkna medelinkomsten
average_income = housing['median_income'].mean()

# Skriv ut medelinkomsten
print(f"Medelinkomsten r: {average_income:.2f}")
```

```python
average_rooms = housing['total_rooms'].mean()

# Skriv ut medelvrdet
print(f"Medelvrdet fr antal rum i husen r: {average_rooms:.2f}")
```

```python
total_population = housing['population'].sum()

# Skriv ut resultatet
print(f"Den totala befolkningen r: {total_population}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ls in data
housing = pd.read_csv("housing.csv")

# Rkna andelen hus nra vatten
ocean_counts = housing['ocean_proximity'].value_counts(normalize=True) * 100

# Skapa ett cirkeldiagram
plt.figure(figsize=(8, 6))
ocean_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['yellow', 'purple', 'steelblue', 'green', 'navy'])

# Anpassa diagrammet
plt.title('Hus nra vatten (ocean_proximity)', fontsize=14)
plt.ylabel('')  # Ta bort standard-etiketten
plt.show()
```
