```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Läs in data
housing = pd.read_csv("housing.csv")

# Skapa en scatter-plot baserad på longitude och latitude
plt.figure(figsize=(10, 8))
plt.scatter(housing['longitude'], housing['latitude'], alpha=0.4, 
            c=housing['median_house_value'], cmap='viridis', s=10)
plt.colorbar(label='Median House Value')
plt.title('Geographical Distribution of Houses')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
missing_values = housing.isnull().sum()

# Skriv ut resultatet
print("Saknade värden i varje kolumn:")
print(missing_values)
```

```python
dyrast = housing.sort_values(by='median_house_value', ascending=False)

# Visa de 10 dyraste husen
print(dyrast.head(10))
```

```python
billigast = housing.sort_values(by='median_house_value', ascending=True)

# Visa de 10 billigast husen
print(billigast.head(10))
```

```python
import pandas as pd

housing = pd.read_csv("housing.csv")

# Beräkna medelinkomsten
average_income = housing['median_income'].mean()

# Skriv ut medelinkomsten
print(f"Medelinkomsten är: {average_income:.2f}")
```

```python
average_rooms = housing['total_rooms'].mean()

# Skriv ut medelvärdet
print(f"Medelvärdet för antal rum i husen är: {average_rooms:.2f}")
```

```python
total_population = housing['population'].sum()

# Skriv ut resultatet
print(f"Den totala befolkningen är: {total_population}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Läs in data
housing = pd.read_csv("housing.csv")

# Räkna andelen hus nära vatten
ocean_counts = housing['ocean_proximity'].value_counts(normalize=True) * 100

# Skapa ett cirkeldiagram
plt.figure(figsize=(8, 6))
ocean_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['yellow', 'purple', 'steelblue', 'green', 'navy'])

# Anpassa diagrammet
plt.title('Hus nära vatten (ocean_proximity)', fontsize=14)
plt.ylabel('')  # Ta bort standard-etiketten
plt.show()
```
